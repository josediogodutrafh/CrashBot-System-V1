```python
# Célula 1: Imports e Carga de Dados

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf

# --- Configurações ---
# Caminho para o banco de dados (ajuste se necessário)
DB_PATH = "../banco de dados/crash_bot_historico.db"

# Configurações de visualização para os gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 7)

# --- Carga dos Dados ---
try:
    with sqlite3.connect(DB_PATH) as conn:
        # Carrega a tabela de multiplicadores em um DataFrame do Pandas
        df = pd.read_sql_query("SELECT id, timestamp, multiplicador FROM multiplicadores_historico", conn)

    # Converte a coluna 'timestamp' para o formato de data/hora
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    # Define o timestamp como o índice do DataFrame para facilitar análises temporais
    df.set_index('timestamp', inplace=True)
    df.sort_index(inplace=True)

    print("✅ Dados carregados com sucesso!")
    print(f"Total de rodadas para análise: {len(df)}")
    
    # Exibe as primeiras 5 linhas para verificação
    display(df.head())
    
except Exception as e:
    print(f"❌ Erro ao carregar dados: {e}")
```
